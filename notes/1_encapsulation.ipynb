### Encapsulation
Allows data and methods to be bundled within a single class. This encapsulation keeps the internal workings, or state, of the object hidden and secure, exposing only necessary aspects through interfaces. Used to organise and manage code efficiently.

### Access Specifiers
`public`, `private`, `protected` help manage the visibility of class members. This enforces ***data hiding***, ensuring class internals reamin secure and inaccessible from outside the class.
- public: full accessibility
- private: restricts it entirely within the class
- protected: permits access to derived classes

### Classes Overview
